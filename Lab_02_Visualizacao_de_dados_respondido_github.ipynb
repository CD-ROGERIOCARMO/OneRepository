## 1. Terminologia

### Indiv√≠duos (amostras) e vari√°veis

ind-var.png

- <span style="color:#6d9eeb"><b>Indiv√≠duo (linha):</b></span> Pessoa/lugar/coisa para a qual os dados s√£o registrados. Tamb√©m chamada de **observa√ß√£o**.

- <span style="color:#ff9900"><b>Vari√°vel (coluna):</b></span> Algo que √© registrado para cada indiv√≠duo. Tamb√©m chamado de **recurso**.

### Tipos de vari√°veis

Existem dois tipos principais de vari√°veis:

- **Num√©rico**: Faz sentido fazer aritm√©tica com os valores.
- **Categ√≥rico**: Os valores se enquadram em categorias, que podem ou n√£o ter alguma _ordem_ para eles.

### Exemplos de vari√°veis ‚Äã‚Äãnum√©ricas

- Sal√°rios dos jogadores da NBA üèÄ.
- Individual: um jogador da NBA.
- Vari√°vel: seu sal√°rio.

- Ganhos brutos do filme üí∞.
- Individual: um filme.
- Vari√°vel: seu rendimento bruto.

- Doses de refor√ßo administradas por dia üíâ.
- Individual: data.
- Vari√°vel: n√∫mero de doses de refor√ßo administradas naquela data.

### Exemplos de vari√°veis ‚Äã‚Äã‚Äã‚Äãcateg√≥ricas

- G√™neros de filmes üé¨.
- Individual: um filme.
- Vari√°vel: seu g√™nero.

- CEPs üè†.
- Pessoa f√≠sica: residente nos EUA.
- Vari√°vel: CEP.
- Apesar de parecerem n√∫meros, os CEPs s√£o categ√≥ricos (a aritm√©tica n√£o faz sentido).

- N√≠vel de experi√™ncia anterior em programa√ß√£o para alunos do DSC 10 üßë‚Äçüéì.
- Individual: aluno do DSC 10.
- Vari√°vel: seu n√≠vel de experi√™ncia anterior em programa√ß√£o, por ex. nenhum, baixo, m√©dio ou alto.
- Existe uma _ordem_ para essas categorias!

### Gr√°ficos de barra

- Os gr√°ficos de barras visualizam a rela√ß√£o entre uma vari√°vel categ√≥rica e uma vari√°vel num√©rica.
- Em um gr√°fico de barras...
- A espessura e o espa√ßamento das barras s√£o arbitr√°rios.
- A ordem dos r√≥tulos categ√≥ricos n√£o importa.
- Para criar um a partir de um DataFrame `df`, use
```
df.plot(
    kind='barh',
    x=categorical_column_name,
    y=numerical_column_name
)
```
- O **"h"** em `'barh'` significa **"horizontal"**.
- √â mais f√°cil ler os r√≥tulos desta forma.
- No gr√°fico anterior, definimos `y='Streams'` mesmo que os streams sejam medidos pelo comprimento do eixo x.

````python
# The bars appear in the opposite order relative to the DataFrame
(charts
 .take(np.arange(10))
 .sort_values(by='streams')
 .plot(kind='barh', x='track_name', y='streams')
);
````

### Quantas m√∫sicas os 15 melhores artistas t√™m entre os 200 melhores?

Primeiro, vamos criar um DataFrame com uma √∫nica coluna que descreve o n√∫mero de m√∫sicas entre as 200 melhores por artista. Isso envolve usar `.groupby` com `.count()`. Como queremos uma linha por artista, agruparemos por `'nomes_artistas'`.

````python
charts
````

````python
songs_per_artist = charts.groupby('artist_names').count()
songs_per_artist
````

Usando `.sort_values` e `.take`, manteremos apenas os 15 melhores artistas. Observe que todas as colunas em `songs_per_artist` cont√™m as mesmas informa√ß√µes (isso √© uma consequ√™ncia do uso de `.count()`).

````python
top_15_artists = (songs_per_artist
                  .sort_values('streams', ascending=False)
                  .take(np.arange(15)))
top_15_artists
````

Usando `.assign` e `.drop`, criaremos uma coluna chamada `'count'` que cont√©m as mesmas informa√ß√µes que as outras 3 colunas cont√™m, e ent√£o `.get` apenas essa coluna (ou equivalentemente, eliminaremos o outras 3 colunas).

````python
# If we give .get a list, it will return a DataFrame instead of a Series!
top_15_artists = (top_15_artists
                  .assign(count=top_15_artists.get('streams'))
                  .get(['count']))
top_15_artists
````

Antes de chamar `.plot(kind='barh', y='count')`, classificaremos `top_15_artists` por `'count'` em ordem **crescente**. Isso ocorre porque, estranhamente, o Python inverte a ordem das linhas ao criar barras em gr√°ficos de barras horizontais.

````python
top_15_artists.sort_values(by='count').plot(kind='barh', y='count');
````

### Gr√°ficos de barras verticais

Para criar um gr√°fico de barras verticais, use `kind='bar'` em vez de `kind='barh'`. No entanto, geralmente s√£o mais dif√≠ceis de ler.

````python
top_15_artists.plot(kind='bar', y='count');
````

## Distribui√ß√µes

### Qual √© a distribui√ß√£o de uma vari√°vel?

- A distribui√ß√£o de uma vari√°vel consiste em todos os valores da vari√°vel que ocorrem nos dados, juntamente com suas frequ√™ncias.
- As distribui√ß√µes ajudam voc√™ a entender:
    - Com que frequ√™ncia uma vari√°vel assume um determinado valor?_
- Ambas as vari√°veis ‚Äã‚Äãcateg√≥ricas e num√©ricas t√™m distribui√ß√µes.

### Vari√°veis ‚Äã‚Äãnum√©ricas

A distribui√ß√£o de uma vari√°vel num√©rica nem sempre pode ser representada com precis√£o por um gr√°fico de barras. Por exemplo, vejamos o n√∫mero de streams de cada uma das 200 m√∫sicas mais populares no Spotify Mundo. üéµ

````python
charts = pd.read_csv('http://www.inf.ufes.br/~vinicius.mota/Intro-IA-CD/04-dataviz/data/regional-global-daily-2024-08-11.csv')
charts = (charts.set_index('rank')
          .assign(million_streams = np.round(charts.get('streams')/1000000, 2))
          .get(['track_name', 'artist_names', 'streams', 'million_streams'])
         )
charts
````

Para ver a distribui√ß√£o do n√∫mero de streams, precisamos agrupar pela coluna `'million_streams'`.

````python
stream_counts = charts.groupby('million_streams').count()
stream_counts = stream_counts.assign(Count=stream_counts.get('track_name')).drop(columns=['track_name', 'artist_names', 'streams'])
stream_counts
````

````python
stream_counts.plot(kind='bar', y='Count', figsize=(15,5));
````

### P√©ssimo de visualizar
- Isso obscurece o fato de que as duas m√∫sicas principais s√£o at√≠picas, com **muito mais streams** do que as outras m√∫sicas.

- O eixo horizontal deve ser num√©rico (como uma reta num√©rica), n√£o categ√≥rico. Deve haver mais espa√ßo entre certas barras do que outras.

## Histogramas

### Histogramas mostram a distribui√ß√£o de vari√°veis ‚Äã‚Äãnum√©ricas *sem normalizar*

Em vez de um gr√°fico de barras, visualizaremos a distribui√ß√£o de uma vari√°vel num√©rica com um **histograma**. Vamos ver como √© um histograma de densidade para `'million_streams'`. O que voc√™ percebe nessa visualiza√ß√£o?

````python
# Ignore o c√≥digo por enquanto.
charts.plot(
    kind='hist',
    y='million_streams',
    bins=np.arange(0, 8, 0.5),
    ec='w'  # borda branca
);
````

Agora √© f√°cil perceber que a maioria das m√∫sicas no top 200 tiveram entre 1 e 2 milh√µes de streams

### Primeira ideia chave por tr√°s dos histogramas: binning üóëÔ∏è

- Binning √© o ato de contar o n√∫mero de valores num√©ricos que se enquadram nos intervalos definidos por dois pontos finais. Esses intervalos s√£o chamados de ‚Äúcaixas‚Äù.
- Um valor cai em uma caixa se for **maior ou igual ao ponto final esquerdo** e **menor que o ponto final direito**.
- [a, b): a est√° inclu√≠do, b n√£o.
- A largura de uma caixa √© sua extremidade direita menos sua extremidade esquerda.


### Personalizando as caixas

- Por padr√£o, o Python agrupar√° seus dados em 10 compartimentos de tamanhos iguais.
- Voc√™ pode especificar outro n√∫mero de compartimentos de tamanhos iguais definindo o argumento opcional `bins` igual a algum outro valor inteiro.
- Voc√™ tamb√©m pode especificar o in√≠cio e os pontos finais do compartimento personalizado definindo `bins` iguais a uma sequ√™ncia de pontos finais do compartimento.
- Pode ser um array `list` ou `numpy`.

````python
charts.plot(
    kind='hist',
    y='million_streams',
    bins=20,
    ec='w'
);
````

````python
charts.plot(
    kind='hist', y='million_streams', density=True,
    bins=[0, 1, 2, 3, 4, 5],
    ec='w'
);
````

Nos dois histogramas acima, o que √© diferente e o que √© igual?

### Observa√ß√µes
- A forma geral de todos os dois histogramas √© a mesma, independentemente dos compartimentos. Esta forma √© chamada de *inclinada √† direita*.
- Mais compartimentos fornecem uma imagem mais precisa e granular da distribui√ß√£o da vari√°vel `'million_streams'`.
- Os valores do eixo $y$ parecem mudar muito quando mudamos os compartimentos.

### Histogramas de densidade mostram a distribui√ß√£o de vari√°veis ‚Äã‚Äãnum√©ricas

Podemos tamb√©m criar histogramas *normalizados* ou histogramas de densidade.

### Ideia chave por tr√°s dos histogramas de densidade: a √°rea total √© 1

- Em um histograma de densidade, o eixo $y$ pode ser dif√≠cil de interpretar, mas foi projetado para dar ao histograma uma propriedade muito boa: $$\textbf{As barras de um histograma de densidade }$$
$$\textbf{t√™m uma √°rea total combinada de 1.}$$
- Isso significa que a √°rea de uma barra √© igual √† propor√ß√£o de todos os pontos de dados que caem nessa caixa.
- Propor√ß√µes e percentagens representam a mesma coisa.
- Uma propor√ß√£o √© um decimal entre 0 e 1, uma porcentagem est√° entre 0\% e 100\%.
- A propor√ß√£o 0,34 significa 34\%.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o dataframe a partir do arquivo CSV
df = pd.read_csv('http://www.inf.ufes.br/~vinicius.mota/Intro-IA-CD/04-dataviz/data/regional-global-daily-2024-08-11.csv')

# Seleciona as 10 m√∫sicas mais populares
top_10_songs = df.sort_values(by='streams', ascending=False).head(10)

# Cria o gr√°fico de barras horizontal
plt.figure(figsize=(10, 5))
plt.barh(top_10_songs['track_name'], top_10_songs['streams'], color='skyblue')
plt.xlabel('Streams')
plt.ylabel('M√∫sica')
plt.title('10 M√∫sicas Mais Populares')
plt.show()
````

### 2 - N√∫mero de m√∫sica no rank dos top artistas

Apresentar um gr√°fico de barra horizontal do n√∫mero de musicas que cada um entre os 10 melhores artistas estiveram no Top Streaming.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o dataframe a partir do arquivo CSV
df = pd.read_csv('http://www.inf.ufes.br/~vinicius.mota/Intro-IA-CD/04-dataviz/data/regional-global-daily-2024-08-11.csv')

# Agrupa por artista e conta o n√∫mero de m√∫sicas
artist_counts = df.groupby('artist_names')['track_name'].nunique()

# Seleciona os 10 melhores artistas
top_10_artists = artist_counts.sort_values(ascending=False).head(10)

# Cria o gr√°fico de barras horizontal
plt.figure(figsize=(10, 5))
plt.barh(top_10_artists.index, top_10_artists.values, color='coral')
plt.xlabel('N√∫mero de M√∫sicas')
plt.ylabel('Artista')
plt.title('N√∫mero de M√∫sicas no Top Streaming por Artista')
plt.show()
````

### 3 -Streams do artista mais tocado

Apresentar um gr√°fico de barra  de quantos streams tiveram cada m√∫sica do artista mais tocado (ver exemplo da se√ß√£o m√∫sicas em destaque)

````python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o dataframe a partir do arquivo CSV
df = pd.read_csv('http://www.inf.ufes.br/~vinicius.mota/Intro-IA-CD/04-dataviz/data/regional-global-daily-2024-08-11.csv')

# Encontra o artista com mais streams
top_artist = df.groupby('artist_names')['streams'].sum().idxmax()

# Filtra o dataframe para o artista mais tocado
top_artist_songs = df[df['artist_names'] == top_artist]

# Cria o gr√°fico de barras
plt.figure(figsize=(10, 5))
plt.bar(top_artist_songs['track_name'], top_artist_songs['streams'], color='mediumseagreen')
plt.xlabel('M√∫sica')
plt.ylabel('Streams')
plt.title('Streams das M√∫sicas de {}'.format(top_artist))
plt.xticks(rotation=45, ha='right')
plt.show()
````
